### EDA on loan_status

```python
## Column: loan_status 
print(loan['loan_status'].dtype)
print(loan['loan_status'].value_counts())

# categories assigned to default: Charged Off, Does not meet the credit policy. Status:Charged Off and Default
# categories assigned to non-default: Fully Paid, Does not meet the credit policy. Status:Fully Paid

retain_li = ['Charged Off', 'Does not meet the credit policy. Status:Charged Off', 'Default', 'Fully Paid',
            'Does not meet the credit policy. Status:Fully Paid']
loan_filtered = loan.ix[loan['loan_status'].isin(retain_li), :]
print('\n----------')
print(loan_filtered['loan_status'].value_counts())

# assign the default categories
default_map = dict()
for i in retain_li:
    if i in ['Charged Off', 'Does not meet the credit policy. Status:Charged Off', 'Default']:
        default_map[i] = 1
    else:
        default_map[i] = 0
print(default_map)

loan_filtered['Default'] = loan_filtered['loan_status'].map(default_map)
print(loan_filtered['Default'].value_counts())
```

```python
## Split the data to training and test set
X = loan_filtered.ix[:, loan_filtered.columns != 'Default']
y = loan_filtered.ix[:, loan_filtered.columns == 'Default']

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=100) # stratified split based on y
print(X_train.shape)
print(y_train.shape)
print(y_train.mean())
print(y_test.mean()) # check the category ratio

loan_train = pd.concat([X_train, y_train], axis=1)
loan_train.index = list(range(len(loan_train)))
print(loan_train.shape)
```

```python
# feature use: 'term', 'int_rate', 'grade', 'sub_grade', 'home_ownership', 'annual_inc', 'purpose', 'dti',

# variable type: term(cat), int_rate(val), grade(cat), sub_grade(cat), home_ownership(cat), annual_inc(val), purpose(cat), dti(value)
# for home_ownership, transform the NONE and ANY to OTHER
# pay extra attention to variables with NAs
# I won't take something like recoveries or late fees into account since they happen after the fact of default
# LC: Lending club
# excluding features that represent the current condition (such as current ..., ... in last 12 months...)
# transsform the categorical variables to one-hot encoding
# scale the numerical variables 
```

```python
### one-hot encoding function 
def one_hot(original_df, filtered_df, col_name):
    from sklearn import preprocessing
    
    col_variable_list = list(original_df[col_name].value_counts().index)
    
    le = preprocessing.LabelEncoder()
    le.fit(col_variable_list)
    
    transformed_label = le.transform(filtered_df[col_name].values).reshape([-1, 1])
    
    en = preprocessing.OneHotEncoder(sparse=False, n_values=len(col_variable_list))
    en.fit(le.transform(col_variable_list).reshape([-1, 1]))
    
    transformed_label_array = en.transform(transformed_label)
    transformed_label_array = pd.DataFrame(transformed_label_array)
    col_name_list = [col_name + '_' + str(x) for x in col_variable_list]
    transformed_label_array.columns = col_name_list
    
    return transformed_label_array
```

```python
### transform the categorical columns into one-hot encoding 
loan_train_filtered = loan_train.ix[:, ['term', 'int_rate', 'grade', 'sub_grade', 'home_ownership', 'annual_inc', 'purpose', 'dti', 'Default']]
print(loan_train_filtered.shape)
print(loan_train_filtered.isnull().any())

for cat_col in ['term', 'grade', 'sub_grade', 'home_ownership', 'purpose']:
    loan_train_filtered = pd.concat([loan_train_filtered, one_hot(loan, loan_train_filtered, cat_col)], axis=1)

print(loan_train_filtered.shape)
```

```python
loan_train_filtered.drop(['term', 'grade', 'sub_grade', 'home_ownership', 'purpose'], axis=1, inplace=True)
print(loan_train_filtered.shape)
```
